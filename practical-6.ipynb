```python
# pra-6 Write a Pandas program to implement following operation
# 6.1 to convert a NumPy array to a Pandas series
import numpy as np
import pandas as pd
a= np.array([1,2,3,4,5])
b = pd.Series(a)
print(b)
```

```python
# 6.2to convert the first column of a DataFrame as a Series
import pandas as pd
df = pd.DataFrame({'A' : [1,2,3],
'B' : [4,5,6]})
s = df.iloc[:,0]
print(s)
```

```python
# 6.3 to create the mean and standard deviation of the data of a given Series

import pandas as pd

s=pd.Series([12,45,78,23,56,89,88])

print("series is:\n",s)
print("\nmean of s:",s.mean())
print("\nstandard deviation of s:",s.std())
```
