```python
# pra-6 Write a Pandas program to implement following operation
# 6.1 to convert a NumPy array to a Pandas series

import numpy as np
import pandas as pd
numpy_array = np.array([1, 2, 3, 4, 5])

pandas_series = pd.Series(numpy_array)
print(numpy_array)
print("converted array :")
print(pandas_series)

```

```python
#6.2  to convert the first column of a DataFrame as a Series

import pandas as pd

data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)

first_column_series = df.iloc[:, 0]

print(first_column_series)
```

```python
#6.3 to create the mean and standard deviation of the data of a given Series

import numpy as np
import pandas as pd
data = pd.Series([1, 2, 3, 4, 5])

mean = np.mean(data)

std_dev = np.std(data)

print(f"Mean: {mean}")
print(f"Standard Deviation: {std_dev}")
```

```python
# 6.4 to sort a given Series
import pandas as pd

data = pd.Series([3, 1, 2, 4])
sorted_series = data.sort_values()
print(sorted_series)
```
