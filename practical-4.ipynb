```python
#Practical - 4 : Write a NumPy program to implement following operation

#Practical - 4.1 to add, subtract, multiply, divide arguments element-wise

import numpy as np

print("Add : ", np.add(1,2))
print("Subtract : ", np.subtract(4,2))
print("Multiply : ", np.multiply(2,2))
print("Divide : ", np.divide(4,2))
```

```python
#Practical - 4.2 to round elements of the array to the nearest integer

import numpy as np

a=np.array([0.10,12.5,11.2,4.9])
print("rounding element : ", np.around(a))
```

```python
#Practical - 4.3 to calculate mean across dimension, in a 2D numpy array

import numpy as np

arr = np.arange(1,10).reshape(3,3)
print("Array : \n",arr)

c = np.mean(arr)
print("\nColoum vise mean : ",c)
r = np.mean(arr,axis=0)
print("Row vise mean : ",r)
```

```python
#Practical - 4.4 to calculate the difference between neighboring elements, element wise of a given array

import numpy as np

arr = np.arange(1,10).reshape(3,3)
print("Array : \n",arr)
print()

diff=np.diff(arr,axis=0)
print("Diff of neighbouring element\n",diff)

diff2=np.diff(arr,axis=1)
print("\nDiff of elements\n",diff2)
```
