```python
#Practical - 3 : Write a NumPy program to implement following operation

#Practical - 3.1 to split an array of 14 elements into 3 arrays, 
#                each with 2, 4, and 8 elements in the original order

import numpy as np

arr = np.arange(1,15)
print("Array : ",arr)
print("Array Splitting : ", np.split(arr,[2,6]))
```

```python
#Practical - 3.2 to stack arrays horizontally (column wise)

import numpy as np

arr1 = np.arange(1,6)
print("Array 1 : ",arr1)

arr2 = np.arange(6,11)
print("Array 2 : ",arr2)

arr3 = np.hstack((arr1,arr2))
print("\nColoum wise : arr3\n",arr3)
```
