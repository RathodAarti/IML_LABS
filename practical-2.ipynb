```python
#Practical - 2.4 to create another shape from an array without changing its data(3*2 to 2*3)
import numpy as np
arr = [1,2,3,4,5,6]
print("Array : ",arr,"\n")
arr1 = np.reshape(arr,(3,2))
print("Array 3x2 : \n",arr1)
arr2 = np.reshape(arr,(2,3))
print("\nArray 2x3 : \n",arr2)
```
